```python
from ultralytics import YOLO
model = YOLO("yolov10n.pt")
image_path = "/Users/anastasia/Desktop/YOLOv10_2/data/images/Cat and dog image.jpeg"
results = model.predict(image_path)

results[0].show()

for box in results[0].boxes:
    print(f"Coordinates: {box.xyxy}, Class: {box.cls}, Confidence: {box.conf}")
```
